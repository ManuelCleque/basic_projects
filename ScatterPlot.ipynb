<h1>Matplotlib en Python</h1><h2> <i>[Using Matplotlib in Python]</i></h2>

Importamos la librer√≠a Matplotlib <i> [Import Matplotlib]

```python
import matplotlib.pyplot as pyplot
from matplotlib.figure import Figure
```

Creamos la data a graficar <i>[Create data]</i>

```python
Lima = ((15,15,18,19,16,35,38,39,39,38,45),(20,22,25,45,20,20,22,40,45,50,45))
Arequipa = ((25,26,24,22,21,22,20,22,35,37,35),(40,45,48,30,20,20,21,22,10,6,8))
Piura = ((20,15,18,21,26,30,32,32,34,38,45),(0,20,5,6,12,15,7,3,5,0,0))
```

Configurar data e imprimir la data <i>[Set and show graphic]</i>

```python
pyplot.xlabel('Age')
pyplot.ylabel('USD per month')
pyplot.title('Money in Streaming Services')
pyplot.scatter(x=Lima[0],y=Lima[1],c='red',marker='s',label="Lima")
pyplot.scatter(x=Arequipa[0],y=Arequipa[1],c='green',marker='o',label="Arequipa")
pyplot.scatter(x=Piura[0],y=Piura[1],c='blue',marker='*',label="Piura")
pyplot.legend(loc="center right")
pyplot.show()
```
